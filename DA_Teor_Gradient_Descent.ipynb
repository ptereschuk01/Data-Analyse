#Y = x2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# параметры
learning_rate = 0.1
x = 4

data = []

# Gradient Descent
for step in range(11):

    fx = x**2
    derivative = 2*x
    change = -learning_rate * derivative
    new_x = x + change

    data.append({
        "step": step,
        "x": x,
        "F(x)": fx,
        "F'(x)": derivative,
        "change": change,
        "new_x": new_x
    })

    x = new_x

# DataFrame
df = pd.DataFrame(data)

print(df)

# --------- ГРАФИК ---------

# значения функции
x_vals = np.linspace(-1, 4.5, 400)
y_vals = x_vals**2

plt.figure(figsize=(8,6))

# парабола
plt.plot(x_vals, y_vals, label="F(x) = x²")

# точки gradient descent
plt.scatter(df["x"], df["F(x)"], s=80)

# соединяем стрелками
for i in range(len(df)-1):
    plt.plot(
        [df["x"][i], df["x"][i+1]],
        [df["F(x)"][i], df["F(x)"][i+1]]
    )

plt.title("Gradient Descent for F(x) = x²")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()

plt.show()
```

#Y = x3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

learning_rate = 0.1
x = 3

data = []

# Gradient Descent
for step in range(10):

    fx = x**2
    derivative = 2*x
    change = -learning_rate * derivative
    new_x = x + change

    data.append([step, x, fx, derivative, change, new_x])

    x = new_x

df = pd.DataFrame(
    data,
    columns=["step","x","F(x)","F'(x)","change","new_x"]
)

print(df)

# -------- ГРАФИК --------

x_vals = np.linspace(0,3,400)
y_vals = x_vals**2

plt.figure(figsize=(10,6))

plt.plot(x_vals, y_vals, label="F(x) = x²")

# точки Gradient Descent
plt.scatter(df["x"], df["F(x)"], s=120)

# линии движения
for i in range(len(df)-1):
    plt.plot(
        [df["x"][i], df["x"][i+1]],
        [df["F(x)"][i], df["F(x)"][i+1]]
    )

plt.xlim(0,3)
plt.ylim(0,9)

plt.title("Gradient Descent (start x = 3)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

learning_rate = 0.1
x = 1

data = []

for step in range(10):
    fx = 1/x
    derivative = -1/(x**2)
    change = -learning_rate * derivative
    new_x = x + change
    data.append([step, x, fx, derivative, change, new_x])
    x = new_x

df = pd.DataFrame(
    data,
    columns=["step", "x", "F(x)", "F'(x)", "change", "new_x"]
)

print(df)

# Значения функции для графика
x_vals = np.linspace(1, 1.6, 400)
y_vals = 1/x_vals

plt.figure(figsize=(10,6))

# График функции
plt.plot(x_vals, y_vals, label="F(x) = 1/x")

# Точки Gradient Descent
plt.scatter(df["x"], df["F(x)"], s=120, color='red')

# Линии движения
for i in range(len(df)-1):
    plt.plot([df["x"][i], df["x"][i+1]],
             [df["F(x)"][i], df["F(x)"][i+1]], color='blue')

# Масштаб
plt.xlim(1, 1.6)
plt.ylim(0.6, 1.0)  # немного выше/ниже для наглядности

plt.title("Gradient Descent (F(x)=1/x) - масштаб X: 1 до 1.6")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()
plt.show()
```
